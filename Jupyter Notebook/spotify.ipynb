```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

Load the Dataset

```python
df = pd.read_csv('data/data_moods.csv')
```

Display Info

```python
df.info()
```

Showing Columns

```python
df.columns
```

Correlations

```python
df.corr(numeric_only = True)
```

Plots

```python
plt.figure(figsize=(16,8))
sns.heatmap(df.corr(numeric_only = True), annot= True, fmt='.2f')
```

```python
df.hist(figsize=(20,20))
plt.title('Histogram plot', size = 15, weight = 'bold')
```

```python
sns.jointplot(x='energy',y='loudness',data=df,kind='scatter')
plt.show()
```

```python
sns.jointplot(x='energy',y='loudness',data=df,kind='hex')
plt.show()
```

```python
sns.jointplot(x='energy', y='loudness', data = df, kind = 'reg')
plt.show()
```

```python
sns.jointplot(x='energy', y='loudness', data = df, kind = 'kde')
plt.show()
```

```python
g = sns.jointplot(data=df, x="energy", y="loudness")
g.plot_joint(sns.kdeplot, color="r", zorder=0, levels=6)
g.plot_marginals(sns.rugplot, color="r", height=-.15, clip_on=False)
```

```python
fig=plt.figure(figsize=(10,10))
sns.regplot(data=df,x='energy',y='loudness')
```

```python
df['Count'] = 1
df_artist = df.groupby('artist')['Count'].sum().reset_index().sort_values(by='Count', ascending = False)
df_artist.head(10)
```

```python
fig = px.bar(df_artist.head(10),x='artist', y='Count',text_auto=True)
fig.update_layout(legend_orientation="h",
                  legend=dict(x=0, y=1, traceorder='normal'),
                  title="Top 10 Artist",
                  margin=dict(l=0, r=0, t=30, b=0))
fig.show()
```

```python
fig=px.pie(df_artist.head(10),values='Count',names='artist',hole=0.4)
fig.update_layout(title='Top 10 Artist',font_size=15,title_x=0.45,annotations=[dict(text='Artist',font_size=25, showarrow=False,height=800,width=700)])
fig.update_traces(textfont_size=15,textinfo='percent')
fig.show()
```
